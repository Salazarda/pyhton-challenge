```python
import pandas as pd
# Create a reference the CSV file desired
csv_path = "budget_data.csv"
file_to_output = "budget_analysis.txt"
# Read the CSV into a Pandas DataFrame
budget_df = pd.read_csv(csv_path)

# Print the first five rows of data to the screen
budget_df.head()

# Count how many months
month_counts = budget_df["Date"].count()
month_counts

net_profit = budget_df['Profit/Losses'].sum()
net_profit

average_change = net_profit / month_counts
average_change

budget_df['Profit/Losses'].min()



AmtChange = budget_df["Profit/Losses"].diff()

GreatestIncrease = budget_df["Profit/Losses"].max()
GreatestDecrease = budget_df["Profit/Losses"].min()
AmtChange.head()

inp_budget_df = pd.DataFrame(budget_df)
inp_budget_df

Gi_Date = budget_df.sort_values('Profit/Losses').tail(1).Date
Gd_Date = budget_df.sort_values('Profit/Losses').head(1).Date

print("Total Months: %s" %(month_counts))
print("Total: %s" %(net_profit))
print("Greatest Increase in Profits: %s %s" %(Gi_Date.to_string(index=False), GreatestIncrease))
print("Greatest Decrease in Profits: %s %s" %(Gd_Date.to_string(index=False), GreatestDecrease))

with open(file_to_output, "w") as txt_file:
    
    txt_file.write("Budget Results")
    txt_file.write("-------------------------")
    txt_file.write(f"Total Profit:  {net_profit}")
    txt_file.write("-------------------------")
    txt_file.write(f"Greatest Increase in Profits: {Gi_Date}")
    txt_file.write("-------------------------")
    txt_file.write(f"Greatest Decrease in Profits: {Gd_Date}")               
    txt_file.write("-------------------------")
```
